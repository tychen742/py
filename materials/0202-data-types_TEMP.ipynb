primitive types vs container (sequences & collections) types

Feature 	"Primitive" Types	Container Types
Purpose	Store single, basic data values.	Store collections of other objects.
Examples	int, float, bool, str (conceptually a single value).	list, tuple, dict, set.
Composition	Cannot be broken down into smaller components of the same type (conceptually).	Composed of other objects, which can be accessed individually (e.g., by index or key).
Mutability	Often immutable (e.g., int, float, bool, str).	Can be mutable (list, dict, set) or immutable (tuple, frozenset).
Methods	Support basic operations (arithmetic, logic).	Have methods for managing their collection of items (e.g., .append(), .keys(), etc.).
