## with Statement for Safe File Handling

Use `with` to manage resources automatically.
For files, it ensures close/cleanup even when errors happen.

```python
from pathlib import Path

path = Path('demo.txt')

with open(path, 'w', encoding='utf-8') as f:
    f.write('hello tooling\n')

with open(path, 'r', encoding='utf-8') as f:
    print(f.read())
```

## __name__ and __main__ Dunder

When a file is executed directly, `__name__` is `'__main__'`.
When imported as a module, `__name__` is the module name.

Use this to separate reusable functions from script-only execution.

```python
def main():
    print('Run script workflow here')

print('__name__ =', __name__)

if __name__ == '__main__':
    main()
```

## List Comprehension and Lambda

These are concise tools for transformation logic in scripts.
- Comprehension: readable filtering/transformation in one expression
- Lambda: lightweight anonymous function for small operations

```python
names = ['alice', 'bob', 'charlie', 'amy']

# list comprehension
starts_with_a = [n for n in names if n.startswith('a')]

# lambda with sorted key
by_length = sorted(names, key=lambda s: len(s))

print(starts_with_a)
print(by_length)
```

```python
### Exercise: Comprehension + Lambda
# Given numbers = [1,2,3,4,5,6], create:
# 1) a list of squares of even numbers using comprehension
# 2) a lambda that doubles a value, then map it over numbers
### Your code starts here.


### Your code ends here.
```

```python
### Solution
numbers = [1, 2, 3, 4, 5, 6]

even_squares = [n * n for n in numbers if n % 2 == 0]
double = lambda x: x * 2
mapped = list(map(double, numbers))

print(even_squares)
print(mapped)
```
