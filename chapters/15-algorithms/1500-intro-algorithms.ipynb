The concept of computer algorithm may be represented as the flowchart below.

         INPUT

    ┌──────────────┐

    └──────┬───────┘
           │

    ┌──────────────┐
    │   STEP  2    │
    └──────┬───────┘
           │

    ┌──────────────┐        ┌─────────────┐
    │   STEP  3    │──YES──▶│  DO  THIS   │

    └──────┬───────┘
           │ NO

````python
import numpy as np
import matplotlib.pyplot as plt
import math

n = np.linspace(1, 10, 300)

def norm(y):
    return y / y.max()

n_factorial = norm(np.exp(np.array([math.lgamma(x + 1) for x in n])))

BG     = "#0a0a14"
CARD   = "#111827"
BORDER = "#1e90ff"

data = [
    ("O(1)",       "Constant Time",       norm(np.ones_like(n)),
     "#00e676", ["Runtime does not change with input size", "Example: array index access"]),
    ("O(log n)",   "Logarithmic Time",    norm(np.log2(n)),
     "#ce93d8", ["Problem size halves each step", "Example: binary search"]),
    ("O(n)",       "Linear Time",         norm(n),
     "#29b6f6", ["Work grows directly with input size", "Example: scanning an array"]),
    ("O(n log n)", "Linearithmic Time",   norm(n * np.log2(n)),
     "#00e676", ["Linear work with logarithmic steps", "Example: merge sort / quicksort"]),
    ("O(n²)",      "Quadratic Time",      norm(n ** 2),
     "#f06292", ["Work grows with square of input", "Example: nested loops / bubble sort"]),
    ("O(2ⁿ)",      "Exponential Time",    norm(2.0 ** n),
     "#ffb74d", ["Work doubles with each new element", "Example: generating subsets"]),
    ("O(n!)",      "Factorial Time",      n_factorial,
     "#ef5350", ["Explodes for even small inputs", "Example: generating permutations"]),
]

fig, axes = plt.subplots(
    7, 2, figsize=(7, 7),
    gridspec_kw={"width_ratios": [1, 2.4], "wspace": 0.0, "hspace": 0.08},
)
fig.patch.set_facecolor(BG)
plt.subplots_adjust(left=0.09, right=0.98, top=0.96, bottom=0.03)

for i, (label, title, y, color, bullets) in enumerate(data):
    ax_p, ax_t = axes[i]

    ax_p.plot(n, y, color=color, linewidth=2)
    ax_p.set_xticks([1, 5, 10])
    ax_p.set_xticklabels(["1", "5", "10"], fontsize=6, color="#aaa")
    ax_p.set_yticks([0, 0.5, 1])
    ax_p.set_yticklabels(["0", "", "1"], fontsize=6, color="#aaa")
    ax_p.tick_params(colors="#555", length=2, pad=1)
    ax_p.set_xlabel("n", fontsize=7, color="#aaa", labelpad=1)
    ax_p.set_ylabel("T", fontsize=7, color="#aaa", labelpad=1)
    ax_p.set_facecolor(CARD)
    for side, vis in [("left", True), ("top", True), ("bottom", True), ("right", False)]:
        ax_p.spines[side].set_visible(vis)
        ax_p.spines[side].set_color(BORDER)
        ax_p.spines[side].set_linewidth(1.2)
    ax_p.text(0.96, 0.96, label, transform=ax_p.transAxes,
              fontsize=7, color=color, ha="right", va="top", fontweight="bold")

    ax_t.set_facecolor(CARD)
    ax_t.set_xticks([])
    ax_t.set_yticks([])
    for side, vis in [("right", True), ("top", True), ("bottom", True), ("left", False)]:
        ax_t.spines[side].set_visible(vis)
        ax_t.spines[side].set_color(BORDER)
        ax_t.spines[side].set_linewidth(1.2)
    ax_t.text(0.04, 0.84, f"{label}  —  {title}",
              fontsize=9.5, fontweight="bold", color=color,
              va="top", transform=ax_t.transAxes)
    for j, bullet in enumerate(bullets):
        ax_t.text(0.04, 0.50 - j * 0.28, f"• {bullet}",
                  fontsize=7.5, va="top", color="#e0e0e0",
                  transform=ax_t.transAxes)

fig.suptitle("7 Must-Know Big-O Complexities", fontsize=13,
             fontweight="bold", color="white")
plt.show()
````
