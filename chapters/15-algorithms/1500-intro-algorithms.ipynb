The concept of computer algorithm may be represented as the flowchart below. 

```text
         INPUT
           │
           ▼
    ┌──────────────┐
    │   STEP  1    │
    └──────┬───────┘
           │
           ▼
    ┌──────────────┐
    │   STEP  2    │
    └──────┬───────┘
           │
           ▼
    ┌──────────────┐        ┌─────────────┐
    │   STEP  3    │──YES──▶│  DO  THIS   │
    │  condition?  │        └─────────────┘
    └──────┬───────┘
           │ NO
           ▼
    ┌──────────────┐
    │   STEP  4    │◀─────────────┐
    └──────┬───────┘              │
           │                      │
           ▼                      │
    ┌──────────────┐              │
    │   loop?      │──────────────┘
    └──────┬───────┘
           │ done
           ▼
         OUTPUT
```

`````python
import numpy as np
import matplotlib.pyplot as plt
import math
from pathlib import Path

Path("../../images").mkdir(exist_ok=True)

n = np.arange(1, 11)

plots = {
    "o1.png": np.ones_like(n),
    "ologn.png": np.log2(n),
    "on.png": n,
    "onlogn.png": n * np.log2(n),
    "on2.png": n**2,
    "o2n.png": 2**n,
    "onfactorial.png": np.array([math.factorial(i) for i in n]),
}

for filename, y in plots.items():
    plt.figure(figsize=(3, 2))
    plt.plot(n, y, linewidth=3)
    plt.xticks([])
    plt.yticks([])
    plt.box(False)
    plt.tight_layout()
    plt.savefig(f"../../images/{filename}", transparent=True)
    plt.close()
`````
