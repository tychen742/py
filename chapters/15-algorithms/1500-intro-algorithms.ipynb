````python
import numpy as np
import matplotlib.pyplot as plt
import math

n = np.linspace(1, 10, 300)

data = [
    ("O(1)",       "Constant Time",       np.ones_like(n),
     "#2ecc71", ["Runtime does not change with input size", "Example: array index access"]),
    ("O(log n)",   "Logarithmic Time",    np.log2(n),
     "#9b59b6", ["Problem size halves each step", "Example: binary search"]),
    ("O(n)",       "Linear Time",         n,
     "#3498db", ["Work grows directly with input size", "Example: scanning an array"]),
    ("O(n log n)", "Linearithmic Time",   n * np.log2(n),
     "#1abc9c", ["Linear work with logarithmic steps", "Example: merge sort / quicksort"]),
    ("O(n²)",      "Quadratic Time",      n ** 2,
     "#e91e8c", ["Work grows with square of input", "Example: nested loops / bubble sort"]),
    ("O(2ⁿ)",      "Exponential Time",    np.clip(2 ** n, 0, n[-1] ** 2),
     "#f39c12", ["Work doubles with each new element", "Example: generating subsets"]),
    ("O(n!)",      "Factorial Time",      np.clip(
        np.array([math.factorial(max(1, round(x))) for x in n]), 0, n[-1] ** 2),
     "#e74c3c", ["Explodes for even small inputs", "Example: generating permutations"]),
]

fig, axes = plt.subplots(
    7, 2, figsize=(9, 12),
    gridspec_kw={"width_ratios": [1, 2.4], "wspace": 0.08, "hspace": 0.55},
)

for i, (label, title, y, color, bullets) in enumerate(data):
    ax_p, ax_t = axes[i]

    # --- plot ---
    ax_p.plot(n, y, color=color, linewidth=2.5)
    ax_p.set_xticks([])
    ax_p.set_yticks([])
    ax_p.set_xlabel("Input Size", fontsize=7, color="#888", labelpad=2)
    ax_p.set_ylabel("Time", fontsize=7, color="#888", labelpad=2)
    for spine in ax_p.spines.values():
        spine.set_visible(False)
    ax_p.spines["bottom"].set_visible(True)
    ax_p.spines["left"].set_visible(True)
    ax_p.spines["bottom"].set_color("#ccc")
    ax_p.spines["left"].set_color("#ccc")
    ax_p.spines["bottom"].set_linewidth(0.8)
    ax_p.spines["left"].set_linewidth(0.8)
    ax_p.set_facecolor("#f9f9f9")

    # --- text ---
    ax_t.axis("off")
    ax_t.text(0.03, 0.82, f"{label}  —  {title}",
              fontsize=11, fontweight="bold", color=color,
              va="top", transform=ax_t.transAxes)
    for j, bullet in enumerate(bullets):
        ax_t.text(0.03, 0.50 - j * 0.28, f"• {bullet}",
                  fontsize=9, va="top", color="#333",
                  transform=ax_t.transAxes)

fig.patch.set_facecolor("white")
plt.suptitle("7 Must-Know Big-O Complexities", fontsize=14,
             fontweight="bold", y=1.01, color="#222")
plt.tight_layout()
plt.show()
````
