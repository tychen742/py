# Dictionaries & Sets

Dictionaries and sets are hash-table-based containers for fast lookup, membership tests, and uniqueness checks.

- Dictionaries map keys to values
- Sets store unique hashable values
- Hashability controls what can be used as a key or set element
- Comprehensions build dictionaries and sets from existing data
- Lookup-heavy programs often depend on these structures
