import numpy as np

# Create an instance (ndarray object)
arr = np.arange(10)     # arr is an instance of ndarray class

# Call methods on the instance using dot notation

```
arr.sum()               # method on instance
arr.mean()              # method on instance
arr.reshape(2, 5)       # method on instance
arr.dtype               # attribute of instance
```

**Key difference:**
`
| Aspect | Function | Instance |
|---|---|---|
| **Call syntax** | `np.function()` | `object.method()` |
| **Belongs to** | Module | Class/Object |
| **Example** | `np.sum(arr)` | `arr.sum()` |
| **Data** | Works on input | Has its own data |
`


**Diagram:**

```
np.arange(10)           ← Function (module-level)
    ↓
   arr (ndarray instance)
    ↓
arr.sum()               ← Method (instance-level)

```
