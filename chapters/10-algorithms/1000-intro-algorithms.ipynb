## Big O Notation and Algorithm Analysis

Big O notation describes how an algorithm's performance scales with input size. It helps us compare algorithms and predict their behavior on large datasets.

**Common Time Complexities:**
- **O(1)** - Constant time: Same performance regardless of input size
- **O(log n)** - Logarithmic: Performance increases slowly with input size
- **O(n)** - Linear: Performance scales directly with input size
- **O(n log n)** - Linearithmic: Common in efficient sorting algorithms
- **O(nÂ²)** - Quadratic: Performance degrades quickly with larger inputs
- **O(2â¿)** - Exponential: Only practical for very small inputs

```python
import time
import random
from typing import List, Optional

# O(1) - Constant time example
def get_first_element(arr: List) -> Optional[int]:
    """Access first element - always takes same time regardless of array size."""
    return arr[0] if arr else None

# O(n) - Linear time example
def find_max(arr: List[int]) -> Optional[int]:
    """Find maximum element - must check every element once."""
    if not arr:
        return None
    max_val = arr[0]
    for num in arr[1:]:
        if num > max_val:
            max_val = num
    return max_val

# O(nÂ²) - Quadratic time example
def bubble_sort(arr: List[int]) -> List[int]:
    """Bubble sort - nested loops make it quadratic."""
    arr = arr.copy()  # Don't modify original
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Test with different sizes to see performance differences
test_data_small = [64, 34, 25, 12, 22, 11, 90]
test_data_large = [random.randint(1, 1000) for _ in range(100)]

print("Small array - Bubble sort:", bubble_sort(test_data_small))
print("Small array - Max element:", find_max(test_data_small))
print("Large array - Max element:", find_max(test_data_large))
```

## Searching Algorithms

Searching algorithms find specific elements in data structures. The choice of algorithm depends on whether the data is sorted and how often you need to search.

```python
# Linear Search - O(n) time complexity
def linear_search(arr: List[int], target: int) -> Optional[int]:
    """Search for target element by checking each position sequentially."""
    for i, element in enumerate(arr):
        if element == target:
            return i
    return None

# Test linear search
numbers = [64, 34, 25, 12, 22, 11, 90, 5]
target = 22

result = linear_search(numbers, target)
if result is not None:
    print(f"Linear search: Found {target} at index {result}")
else:
    print(f"Linear search: {target} not found")

# Linear search works on unsorted data
print(f"Searching for 11:", linear_search(numbers, 11))
print(f"Searching for 99:", linear_search(numbers, 99))
```

```python
# Binary Search - O(log n) time complexity
# Requires sorted data!
def binary_search(arr: List[int], target: int) -> Optional[int]:
    """Search for target in sorted array by repeatedly halving search space."""
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return None

# Binary search requires sorted data
sorted_numbers = sorted(numbers)
print(f"Sorted array: {sorted_numbers}")

result = binary_search(sorted_numbers, target)
if result is not None:
    print(f"Binary search: Found {target} at index {result}")
else:
    print(f"Binary search: {target} not found")

# Demonstrate efficiency difference
large_sorted = sorted([random.randint(1, 10000) for _ in range(1000)])
target_large = large_sorted[500]  # Pick a target we know exists

start_time = time.time()
linear_result = linear_search(large_sorted, target_large)
linear_time = time.time() - start_time

start_time = time.time()
binary_result = binary_search(large_sorted, target_large)
binary_time = time.time() - start_time

print(f"\nPerformance on 1000 elements:")
print(f"Linear search: {linear_time:.6f} seconds")
print(f"Binary search: {binary_time:.6f} seconds")
print(f"Binary search is ~{linear_time/binary_time:.1f}x faster")
```

## Sorting Algorithms

Sorting algorithms arrange data in a specific order. Different algorithms have different time complexities and are suitable for different scenarios.

```python
# Merge Sort - O(n log n) time complexity
# Divide and conquer approach
def merge_sort(arr: List[int]) -> List[int]:
    """Efficient sorting using divide-and-conquer strategy."""
    if len(arr) <= 1:
        return arr
    
    # Divide
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    # Conquer (merge)
    return merge(left, right)

def merge(left: List[int], right: List[int]) -> List[int]:
    """Merge two sorted arrays into one sorted array."""
    result = []
    i = j = 0
    
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # Add remaining elements
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result

# Test merge sort
test_array = [64, 34, 25, 12, 22, 11, 90, 88, 76, 50, 42]
print(f"Original: {test_array}")
sorted_array = merge_sort(test_array)
print(f"Merge sorted: {sorted_array}")

# Compare sorting algorithms performance
import time

def time_sort_algorithm(sort_func, data, name):
    """Time how long a sorting algorithm takes."""
    start = time.time()
    result = sort_func(data.copy())
    end = time.time()
    print(f"{name}: {end - start:.6f} seconds")
    return result

# Generate larger test data
large_data = [random.randint(1, 1000) for _ in range(500)]

print("\nSorting 500 random numbers:")
time_sort_algorithm(bubble_sort, large_data, "Bubble Sort O(nÂ²)")
time_sort_algorithm(merge_sort, large_data, "Merge Sort O(n log n)")
time_sort_algorithm(sorted, large_data, "Python's built-in (Timsort)")
```

```python
# Quick Sort - Average O(n log n), Worst O(nÂ²)
# Another divide-and-conquer approach
def quicksort(arr: List[int]) -> List[int]:
    """Sort array using quicksort algorithm with random pivot."""
    if len(arr) <= 1:
        return arr
    
    # Choose random pivot to avoid worst-case on sorted data
    pivot_index = random.randint(0, len(arr) - 1)
    pivot = arr[pivot_index]
    
    # Partition
    less = [x for x in arr if x < pivot]
    equal = [x for x in arr if x == pivot]
    greater = [x for x in arr if x > pivot]
    
    # Recursively sort partitions
    return quicksort(less) + equal + quicksort(greater)

# Test quicksort
print(f"\nQuicksort test:")
test_data = [3, 6, 8, 10, 1, 2, 1]
print(f"Original: {test_data}")
print(f"Quicksorted: {quicksort(test_data)}")

# In-place version (more memory efficient)
def quicksort_inplace(arr: List[int], low: int = 0, high: int = None) -> None:
    """In-place quicksort to save memory."""
    if high is None:
        high = len(arr) - 1
    
    if low < high:
        # Partition and get pivot position
        pivot_pos = partition(arr, low, high)
        
        # Recursively sort elements before and after partition
        quicksort_inplace(arr, low, pivot_pos - 1)
        quicksort_inplace(arr, pivot_pos + 1, high)

def partition(arr: List[int], low: int, high: int) -> int:
    """Partition array around pivot element."""
    pivot = arr[high]  # Choose last element as pivot
    i = low - 1  # Index of smaller element
    
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

# Test in-place quicksort
inplace_test = [64, 34, 25, 12, 22, 11, 90]
print(f"\nBefore in-place quicksort: {inplace_test}")
quicksort_inplace(inplace_test)
print(f"After in-place quicksort: {inplace_test}")
```

## Graph Algorithms

Graphs represent relationships between objects. They consist of vertices (nodes) and edges (connections). Graph algorithms are essential for problems involving networks, social relationships, maps, and more.

**Graph Representations:**
- **Adjacency List**: Each vertex stores a list of its neighbors
- **Adjacency Matrix**: 2D matrix where entry (i,j) indicates edge from vertex i to j

**Common Graph Problems:**
- Finding paths between vertices
- Detecting cycles
- Finding shortest paths
- Identifying connected components

```python
from collections import defaultdict, deque
from typing import Dict, List, Set

class Graph:
    """Simple graph implementation using adjacency list."""
    
    def __init__(self):
        self.graph = defaultdict(list)
    
    def add_edge(self, u, v):
        """Add an edge from vertex u to vertex v."""
        self.graph[u].append(v)
    
    def add_undirected_edge(self, u, v):
        """Add an undirected edge between u and v."""
        self.graph[u].append(v)
        self.graph[v].append(u)
    
    def get_vertices(self):
        """Get all vertices in the graph."""
        vertices = set(self.graph.keys())
        for neighbors in self.graph.values():
            vertices.update(neighbors)
        return list(vertices)
    
    def __str__(self):
        result = []
        for vertex in sorted(self.graph.keys()):
            neighbors = ', '.join(map(str, sorted(self.graph[vertex])))
            result.append(f"{vertex}: [{neighbors}]")
        return '\n'.join(result)

# Create a sample graph
g = Graph()
g.add_undirected_edge(0, 1)
g.add_undirected_edge(0, 2)
g.add_undirected_edge(1, 2)
g.add_undirected_edge(2, 3)
g.add_undirected_edge(3, 4)
g.add_undirected_edge(4, 5)
g.add_undirected_edge(3, 5)

print("Graph representation:")
print(g)
print(f"\nVertices: {sorted(g.get_vertices())}")
```

```python
# Depth-First Search (DFS) - uses stack (recursion)
def dfs(graph: Graph, start: int, visited: Set[int] = None) -> List[int]:
    """Depth-first traversal of graph starting from given vertex."""
    if visited is None:
        visited = set()
    
    path = []
    
    if start not in visited:
        visited.add(start)
        path.append(start)
        
        # Recursively visit all unvisited neighbors
        for neighbor in sorted(graph.graph[start]):  # Sort for consistent output
            if neighbor not in visited:
                path.extend(dfs(graph, neighbor, visited))
    
    return path

# Breadth-First Search (BFS) - uses queue
def bfs(graph: Graph, start: int) -> List[int]:
    """Breadth-first traversal of graph starting from given vertex."""
    visited = set()
    queue = deque([start])
    path = []
    
    while queue:
        vertex = queue.popleft()
        
        if vertex not in visited:
            visited.add(vertex)
            path.append(vertex)
            
            # Add all unvisited neighbors to queue
            for neighbor in sorted(graph.graph[vertex]):  # Sort for consistent output
                if neighbor not in visited:
                    queue.append(neighbor)
    
    return path

# Test graph traversal algorithms
print("\nGraph traversal starting from vertex 0:")
dfs_path = dfs(g, 0)
bfs_path = bfs(g, 0)

print(f"DFS path: {dfs_path}")
print(f"BFS path: {bfs_path}")

# Find path between two vertices using BFS
def find_path(graph: Graph, start: int, end: int) -> Optional[List[int]]:
    """Find shortest path between start and end vertices using BFS."""
    if start == end:
        return [start]
    
    visited = {start}
    queue = deque([(start, [start])])
    
    while queue:
        vertex, path = queue.popleft()
        
        for neighbor in graph.graph[vertex]:
            if neighbor not in visited:
                new_path = path + [neighbor]
                
                if neighbor == end:
                    return new_path
                
                visited.add(neighbor)
                queue.append((neighbor, new_path))
    
    return None  # No path found

# Test pathfinding
path_0_to_5 = find_path(g, 0, 5)
path_1_to_4 = find_path(g, 1, 4)

print(f"\nPath from 0 to 5: {path_0_to_5}")
print(f"Path from 1 to 4: {path_1_to_4}")
```

## Dynamic Programming

Dynamic Programming (DP) solves complex problems by breaking them into overlapping subproblems and storing the results to avoid redundant calculations. It's particularly useful for optimization problems.

**Key Characteristics:**
1. **Optimal Substructure**: Solution can be constructed from optimal solutions of subproblems
2. **Overlapping Subproblems**: Same subproblems are solved multiple times
3. **Memoization**: Store results of subproblems to avoid recomputation

**Common DP Problems:**
- Fibonacci sequence
- Longest common subsequence
- Knapsack problem
- Edit distance

```python
# Classic DP example: Fibonacci sequence

# Naive recursive approach - O(2^n) time complexity
def fibonacci_naive(n: int) -> int:
    """Naive recursive Fibonacci - very slow for large n."""
    if n <= 1:
        return n
    return fibonacci_naive(n-1) + fibonacci_naive(n-2)

# Dynamic Programming approach with memoization - O(n) time
def fibonacci_memo(n: int, memo: Dict[int, int] = None) -> int:
    """Fibonacci with memoization to avoid redundant calculations."""
    if memo is None:
        memo = {}
    
    if n in memo:
        return memo[n]
    
    if n <= 1:
        return n
    
    memo[n] = fibonacci_memo(n-1, memo) + fibonacci_memo(n-2, memo)
    return memo[n]

# Bottom-up DP approach - O(n) time, O(1) space
def fibonacci_bottom_up(n: int) -> int:
    """Bottom-up Fibonacci using iteration - most efficient."""
    if n <= 1:
        return n
    
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    
    return b

# Compare performance
print("Fibonacci calculations:")

for i in range(10):
    print(f"fib({i}) = {fibonacci_memo(i)}")

# Time comparison for larger numbers
import time

n = 35
print(f"\nCalculating Fibonacci({n}):")

# Skip naive for large n as it's too slow
if n <= 30:
    start = time.time()
    result_naive = fibonacci_naive(n)
    naive_time = time.time() - start
    print(f"Naive: {result_naive} in {naive_time:.4f} seconds")

start = time.time()
result_memo = fibonacci_memo(n)
memo_time = time.time() - start
print(f"Memoized: {result_memo} in {memo_time:.6f} seconds")

start = time.time()
result_bottom_up = fibonacci_bottom_up(n)
bottom_up_time = time.time() - start
print(f"Bottom-up: {result_bottom_up} in {bottom_up_time:.6f} seconds")
```

```python
# 0/1 Knapsack Problem - Classic DP example
def knapsack(weights: List[int], values: List[int], capacity: int) -> int:
    """
    Solve 0/1 knapsack problem using dynamic programming.
    
    Args:
        weights: List of item weights
        values: List of item values
        capacity: Maximum weight capacity
    
    Returns:
        Maximum value that can be obtained
    """
    n = len(weights)
    
    # Create DP table: dp[i][w] = max value using first i items with weight <= w
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]
    
    # Fill the table bottom-up
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            # Don't include current item
            dp[i][w] = dp[i-1][w]
            
            # Include current item if it fits
            if weights[i-1] <= w:
                include_value = values[i-1] + dp[i-1][w - weights[i-1]]
                dp[i][w] = max(dp[i][w], include_value)
    
    return dp[n][capacity]

def knapsack_with_items(weights: List[int], values: List[int], capacity: int) -> tuple:
    """
    Solve knapsack and return both max value and selected items.
    """
    n = len(weights)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]
    
    # Fill DP table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            dp[i][w] = dp[i-1][w]
            if weights[i-1] <= w:
                include_value = values[i-1] + dp[i-1][w - weights[i-1]]
                dp[i][w] = max(dp[i][w], include_value)
    
    # Backtrack to find selected items
    selected_items = []
    w = capacity
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i-1][w]:
            selected_items.append(i-1)  # Item index
            w -= weights[i-1]
    
    return dp[n][capacity], selected_items

# Test knapsack problem
item_weights = [2, 1, 3, 2, 5]
item_values = [3, 2, 4, 2, 6]
knapsack_capacity = 5

max_value = knapsack(item_weights, item_values, knapsack_capacity)
max_value_with_items, selected = knapsack_with_items(item_weights, item_values, knapsack_capacity)

print("\nKnapsack Problem:")
print(f"Items: {list(zip(item_weights, item_values))} (weight, value)")
print(f"Capacity: {knapsack_capacity}")
print(f"Maximum value: {max_value}")
print(f"Selected items (indices): {selected}")

# Show selected items details
print("Selected items details:")
total_weight = 0
for idx in selected:
    print(f"  Item {idx}: weight={item_weights[idx]}, value={item_values[idx]}")
    total_weight += item_weights[idx]
print(f"Total weight: {total_weight}/{knapsack_capacity}")
```

## Greedy Algorithms

Greedy algorithms make locally optimal choices at each step, hoping to find a global optimum. They don't always produce optimal solutions, but work well for certain problems and are often simpler to implement than DP solutions.

**Characteristics:**
- Make the best choice at each step
- Never reconsider previous choices
- Work well when local optima lead to global optimum
- Generally faster than exhaustive search methods

**Common Greedy Problems:**
- Activity selection
- Fractional knapsack
- Huffman coding
- Minimum spanning tree

```python
# Activity Selection Problem - Classic Greedy Example
def activity_selection(start_times: List[int], end_times: List[int]) -> List[int]:
    """
    Select maximum number of non-overlapping activities.
    Greedy strategy: Always pick the activity that ends earliest.
    """
    n = len(start_times)
    
    # Create list of activities with their indices
    activities = [(start_times[i], end_times[i], i) for i in range(n)]
    
    # Sort by end time (greedy choice)
    activities.sort(key=lambda x: x[1])
    
    selected = [activities[0][2]]  # Select first activity
    last_end_time = activities[0][1]
    
    # Select activities that don't overlap with previously selected
    for start, end, idx in activities[1:]:
        if start >= last_end_time:  # No overlap
            selected.append(idx)
            last_end_time = end
    
    return selected

# Test activity selection
start = [1, 3, 0, 5, 8, 5]
end =   [2, 4, 6, 7, 9, 9]

selected_activities = activity_selection(start, end)

print("Activity Selection Problem:")
print("Activities (start, end):")
for i in range(len(start)):
    marker = " â† SELECTED" if i in selected_activities else ""
    print(f"  Activity {i}: ({start[i]}, {end[i]}){marker}")

print(f"\nMaximum activities selected: {len(selected_activities)}")
print(f"Selected activity indices: {selected_activities}")

# Fractional Knapsack - Another Greedy Example
def fractional_knapsack(weights: List[int], values: List[int], capacity: int) -> float:
    """
    Solve fractional knapsack using greedy approach.
    Strategy: Sort by value/weight ratio and take items greedily.
    """
    n = len(weights)
    
    # Calculate value-to-weight ratios
    items = [(values[i]/weights[i], weights[i], values[i], i) for i in range(n)]
    
    # Sort by ratio in descending order (greedy choice)
    items.sort(reverse=True)
    
    total_value = 0.0
    remaining_capacity = capacity
    selected_items = []
    
    for ratio, weight, value, idx in items:
        if remaining_capacity >= weight:
            # Take whole item
            total_value += value
            remaining_capacity -= weight
            selected_items.append((idx, 1.0))  # (item_index, fraction_taken)
        elif remaining_capacity > 0:
            # Take fraction of item
            fraction = remaining_capacity / weight
            total_value += value * fraction
            selected_items.append((idx, fraction))
            remaining_capacity = 0
            break
    
    return total_value, selected_items

# Test fractional knapsack
frac_weights = [2, 1, 3, 2]
frac_values = [3, 2, 4, 2]
frac_capacity = 5

max_frac_value, frac_selected = fractional_knapsack(frac_weights, frac_values, frac_capacity)

print("\nFractional Knapsack Problem:")
print("Items (weight, value, ratio):")
for i in range(len(frac_weights)):
    ratio = frac_values[i] / frac_weights[i]
    print(f"  Item {i}: ({frac_weights[i]}, {frac_values[i]}, {ratio:.2f})")

print(f"\nCapacity: {frac_capacity}")
print(f"Maximum value: {max_frac_value:.2f}")
print("Selected items:")
for idx, fraction in frac_selected:
    print(f"  Item {idx}: {fraction:.2f} (fraction taken)")
```

## Algorithm Selection Guide

Choosing the right algorithm depends on your specific problem constraints and requirements.

### Time vs Space Trade-offs
| Algorithm Type | Time Complexity | Space Complexity | When to Use |
|---|---|---|---|
| Linear Search | O(n) | O(1) | Small datasets, unsorted data |
| Binary Search | O(log n) | O(1) | Large sorted datasets |
| Bubble Sort | O(nÂ²) | O(1) | Teaching purposes only |
| Merge Sort | O(n log n) | O(n) | Stable sort needed, consistent performance |
| Quick Sort | O(n log n) avg | O(log n) | General purpose, good average case |
| Hash Table | O(1) avg | O(n) | Fast lookups with unique keys |

### Problem-Solving Strategies

**Use Dynamic Programming when:**
- Problem has optimal substructure
- Overlapping subproblems exist
- Need guaranteed optimal solution

**Use Greedy Algorithms when:**
- Local optimal choices lead to global optimum
- Problem exhibits greedy choice property
- Fast approximation is acceptable

**Use Graph Algorithms when:**
- Dealing with relationships/networks
- Need to find paths or connectivity
- Working with hierarchical data

- Depth-First Search (DFS): Exploring as deep as possible
- Breadth-First Search (BFS): Level-by-level exploration
- Path finding: Shortest path between vertices

**Dynamic Programming**: Solving complex optimization problems
- Memoization: Storing subproblem results
- Bottom-up approach: Building solutions incrementally
- Applications: Fibonacci, knapsack, longest common subsequence

**Greedy Algorithms**: Making locally optimal choices
- Activity selection: Choosing non-overlapping activities
- Fractional knapsack: Maximizing value with weight constraints
- When greedy works: Problems with greedy choice property

### Algorithm Selection Principles

1. **Understand your data**: Size, sorted status, update frequency
2. **Know your constraints**: Time limits, memory restrictions
3. **Consider trade-offs**: Simplicity vs efficiency, time vs space
4. **Profile and measure**: Don't optimize prematurely
5. **Use proven solutions**: Leverage well-tested algorithms
