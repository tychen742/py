# Functional Programming

**Learning goals** — By the end of this chapter you will be able to:

- Use **lambda functions** with `map()`, `filter()`, and `sorted()` as concise key functions
- Write **list, dictionary, and set comprehensions** with filter conditions
- Build and apply **decorators** to add logging, timing, or validation without modifying functions
- Apply **recursion** to problems with a natural self-similar structure
- Use **context managers** (`with` blocks and `@contextmanager`) to manage resources safely

Functional-style Python isn't a separate paradigm you switch into — it's a toolkit of concise, composable patterns that make everyday code shorter and clearer.
