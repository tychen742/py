## Python History

Python was created by Guido van Rossum in the late 1980s and first released publicly in 1991. It was designed to emphasize readability and a clean, expressive syntax. The language is open source and guided by the Python Software Foundation.

**Key milestones**
- 1991: First public release (Python 0.9.0).
- 2000: Python 2.0 introduces list comprehensions and garbage collection.
- 2008: Python 3.0 releases with backward-incompatible cleanups (e.g., print as a function).
- 2020: Python 2 reaches end of life; Python 3 becomes the standard.
