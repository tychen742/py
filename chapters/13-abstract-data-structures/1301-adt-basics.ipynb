## Abstract Types and Implementations

An **abstract data type** defines behavior. It says what operations are supported and what each operation means. A **concrete implementation** says how values are stored and how each operation is performed.

| Abstract idea | Required behavior | Possible Python implementation |
| --- | --- | --- |
| Stack | Add and remove from the same end | `list` |
| Queue | Add at the back, remove from the front | `collections.deque` |
| Set | Store unique values and test membership | `set` |
| Map | Associate keys with values | `dict` |
| Graph | Store nodes and relationships | `dict[str, list[str]]` |

The important design question is not which container looks familiar. The important question is which operations the problem needs.

````python
names_list = ["alice", "bob", "charlie"]
names_set = {"alice", "bob", "charlie"}

print("bob" in names_list)
print("bob" in names_set)
````

````python
from collections import deque

# Two implementations of the same LIFO (last-in, first-out) behavior

# list-backed
stack_list = []
stack_list.append("first")
stack_list.append("second")
stack_list.append("third")
print("list  pop:", stack_list.pop())    # third

# deque-backed
stack_deque = deque()
stack_deque.append("first")
stack_deque.append("second")
stack_deque.append("third")
print("deque pop:", stack_deque.pop())   # third

# The caller sees identical behavior.
# The choice of implementation affects performance, not correctness.
````

## Operation Tradeoffs

A data structure is a tradeoff. One structure may make lookup easy but insertion expensive. Another may make insertion easy but searching slower.

The four operations to reason about are **access**, **search**, **insert**, and **delete**.

| Operation | `list` | `set` / `dict` | `deque` |
| --- | --- | --- | --- |
| Access by index | O(1) | — | O(n) |
| Search for value | O(n) | O(1) avg | O(n) |
| Insert at back | O(1) amortized | O(1) avg | O(1) |
| Insert at front | O(n) | — | O(1) |
| Delete by value | O(n) | O(1) avg | O(n) |

Use this table to match the dominant operation in your workload to the right structure:

| Dominant need | Best choice | Why |
| --- | --- | --- |
| Ordered history, access by position | `list` | O(1) index access |
| Fast membership or deduplication | `set` | O(1) avg search |
| Key-value lookup | `dict` | O(1) avg lookup |
| Add or remove from both ends | `deque` | O(1) at front and back |
| Hierarchical containment | tree | O(log n) search when balanced |
| Many-to-many relationships | graph | Depends on traversal |

Before coding, write down the operations your program will perform most often.


````python
import time

n = 100_000
data_list = list(range(n))
data_set = set(data_list)
target = n - 1  # worst case for list: last element

start = time.perf_counter()
_ = target in data_list
list_time = time.perf_counter() - start

start = time.perf_counter()
_ = target in data_set
set_time = time.perf_counter() - start

print(f"list  (O(n)):     {list_time:.6f} s   — scans up to {n:,} items")
print(f"set   (O(1) avg): {set_time:.6f} s   — jumps via hash")
````

:::{note} Operation Costs and Big O

You have seen **O(1)** and **O(n)** in Chapters 6 and 7. Here is a recap, plus **O(log n)**, which appears with trees.

| Notation | Name | Meaning | ADT example |
| --- | --- | --- | --- |
| O(1) | Constant time | Same speed regardless of size | `dict` key lookup, `set` membership |
| O(n) | Linear time | Scales with collection size | `list` search, linear scan |
| O(log n) | Logarithmic time | Halves the search space each step | Binary search tree lookup |

A binary search tree can find any value in O(log n) steps: start at the root, go left if the target is smaller, go right if larger, and repeat. After each step, half the remaining nodes are eliminated — the same principle behind binary search on a sorted list.

:::

## Summary

Abstract data types focus on behavior. Concrete implementations focus on storage and execution. Good data-structure choices start with the operations the program needs most.
