## Section 9.2 — Regular Expressions


### Question 4


```python
### EXERCISE: Regex Escape Sequences
s = "Price: $19.95, code=A_7, spaces here"
### Your code starts here:


### Your code ends here.
```

```python
### Solution
s = "Price: $19.95, code=A_7, spaces here"
print(re.findall(r'\d+', s))
print(re.findall(r'\w+', s))
print(re.findall(r'\$', s))
print(re.findall(r'\.', s))
```

### Question 5


```python
### EXERCISE: Capture Groups
line = "name=Bob,age=27,dept=Sales"
### Your code starts here:


### Your code ends here.
```

```python
### Solution
line = "name=Bob,age=27,dept=Sales"
m = re.search(r'name=([A-Za-z]+),age=(\d+),dept=([A-Za-z]+)', line)
print(m.group(1))
print(m.group(2))
print(m.group(3))
```

### Question 6


```python
### EXERCISE: Full String Validation
codes = ['CS-101', 'MATH-240', 'CS101', 'EE-7']
### Your code starts here:


### Your code ends here.
```

```python
### Solution
codes = ['CS-101', 'MATH-240', 'CS101', 'EE-7']
pattern = r'[A-Z]{2,4}-\d{3}'
for code_str in codes:
    print(code_str, bool(re.fullmatch(pattern, code_str)))
```

### Question 8


```python
### EXERCISE: Word Frequency Counter
text = "the cat sat on the mat and the cat sat"
### Your code starts here:


### Your code ends here.
```

```python
### Solution
text = "the cat sat on the mat and the cat sat"
counter = {}
for word in text.split():
    counter[word] = counter.get(word, 0) + 1
items = sorted(counter.items(), key=lambda x: x[1], reverse=True)
for word, freq in items[:3]:
    print(freq, word, sep='\t')
```

### Question 9


```python
### EXERCISE: Function with Optional Parameter
word_counter = {'the': 3, 'cat': 2, 'sat': 2, 'on': 1, 'mat': 1, 'and': 1}
### Your code starts here:


### Your code ends here.
```

```python
### Solution
def print_least_common(word_counter, num=3):
    items = sorted(word_counter.items(), key=lambda x: x[1])
    for word, freq in items[:num]:
        print(freq, word, sep='\t')

word_counter = {'the': 3, 'cat': 2, 'sat': 2, 'on': 1, 'mat': 1, 'and': 1}
print_least_common(word_counter)
print()
print_least_common(word_counter, 5)
```
