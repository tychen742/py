## Section 12.1 — OOP Design & Methods

### Question 1

```python
### Exercise: Define a Class with Methods
#   Define a class `Student` with:
#   1. `__init__(self, name, grade)` — store both as instance attributes.
#   2. `__str__` — returns a string like 'Student(Alice, 88)'.
#   3. `is_passing(self)` — returns True if grade >= 60, False otherwise.
#   Test:
#     s = Student('Alice', 88)
#     print(s)              -> Student(Alice, 88)
#     print(s.is_passing()) -> True
#     print(Student('Bob', 55).is_passing()) -> False
### Your code starts here.



### Your code ends here.
```

```python
### Solution
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __str__(self):
        return f'Student({self.name}, {self.grade})'

    def is_passing(self):
        return self.grade >= 60

s = Student('Alice', 88)
print(s)                               # Student(Alice, 88)
print(s.is_passing())                  # True
print(Student('Bob', 55).is_passing()) # False
```

### Question 2

```python
### Exercise: Properties and __repr__
#   Define a class `Circle` with:
#   1. `__init__(self, radius)` — store radius as a private attribute `_radius`.
#      Raise a `ValueError` if radius <= 0.
#   2. A `radius` property that returns `_radius`.
#   3. `area(self)` — returns π * radius² (use math.pi).
#   4. `__repr__` — returns a string like 'Circle(radius=5)'.
#   Test:
#     c = Circle(5)
#     print(c)         -> Circle(radius=5)
#     print(c.radius)  -> 5
#     print(round(c.area(), 4)) -> 78.5398
#     Circle(-1)       -> raises ValueError
### Your code starts here.



### Your code ends here.
```

```python
### Solution
import math

class Circle:
    def __init__(self, radius):
        if radius <= 0:
            raise ValueError(f'Radius must be positive, got {radius}')
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    def area(self):
        return math.pi * self._radius ** 2

    def __repr__(self):
        return f'Circle(radius={self._radius})'

c = Circle(5)
print(c)                  # Circle(radius=5)
print(c.radius)           # 5
print(round(c.area(), 4)) # 78.5398

try:
    Circle(-1)
except ValueError as e:
    print(f'ValueError: {e}')
```

## Section 12.2 — OOP Pillars

### Question 3

```python
### Exercise: Inheritance and Method Overriding
#   1. Define a base class `Animal` with:
#        - `__init__(self, name)` — store name.
#        - `speak(self)` — returns the string '{name} makes a sound.'.
#   2. Define `Dog(Animal)` that overrides `speak` to return '{name} barks.'.
#   3. Define `Cat(Animal)` that overrides `speak` to return '{name} meows.'.
#   4. Write a function `make_noise(animals)` that iterates over a list of
#      Animal objects and prints the result of speak() for each one.
#   Test:
#     make_noise([Dog('Rex'), Cat('Whiskers'), Animal('Parrot')])
#     -> Rex barks.
#        Whiskers meows.
#        Parrot makes a sound.
### Your code starts here.



### Your code ends here.
```

```python
### Solution
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f'{self.name} makes a sound.'


class Dog(Animal):
    def speak(self):
        return f'{self.name} barks.'


class Cat(Animal):
    def speak(self):
        return f'{self.name} meows.'


def make_noise(animals):
    for animal in animals:
        print(animal.speak())


make_noise([Dog('Rex'), Cat('Whiskers'), Animal('Parrot')])
# Rex barks.
# Whiskers meows.
# Parrot makes a sound.
```

### Question 4

```python
### Exercise: Encapsulation with a BankAccount
#   Define a class `BankAccount` with:
#   1. `__init__(self, owner, balance=0.0)` — store owner; store balance
#      as a private attribute `_balance`. Raise `ValueError` if balance < 0.
#   2. A `balance` property that returns `_balance`.
#   3. `deposit(self, amount)` — adds amount to `_balance`.
#      Raise `ValueError` if amount <= 0.
#   4. `withdraw(self, amount)` — subtracts amount from `_balance`.
#      Raise `ValueError` if amount <= 0 or amount > _balance.
#   5. `__str__` — returns 'BankAccount(owner=<owner>, balance=$<balance:.2f>)'.
#   Test:
#     acct = BankAccount('Alex', 100.0)
#     acct.deposit(50)
#     acct.withdraw(30)
#     print(acct)          -> BankAccount(owner=Alex, balance=$120.00)
#     print(acct.balance)  -> 120.0
#     acct.withdraw(200)   -> raises ValueError
### Your code starts here.



### Your code ends here.
```

```python
### Solution
class BankAccount:
    def __init__(self, owner, balance=0.0):
        if balance < 0:
            raise ValueError('Initial balance cannot be negative')
        self.owner = owner
        self._balance = balance

    @property
    def balance(self):
        return self._balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError('Deposit amount must be positive')
        self._balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError('Withdrawal amount must be positive')
        if amount > self._balance:
            raise ValueError('Insufficient funds')
        self._balance -= amount

    def __str__(self):
        return f'BankAccount(owner={self.owner}, balance=${self._balance:.2f})'


acct = BankAccount('Alex', 100.0)
acct.deposit(50)
acct.withdraw(30)
print(acct)         # BankAccount(owner=Alex, balance=$120.00)
print(acct.balance) # 120.0

try:
    acct.withdraw(200)
except ValueError as e:
    print(f'ValueError: {e}')
```
