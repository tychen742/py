### Question 4:
The subtract function returns a new dictionary containing all keys from d1 that are not in d2. What is the output of this code?
```python
def subtract(d1, d2):
    res = {}
    for key in d1:
        if key not in d2:
            res[key] = d1[key]
    return res

book_words = {'a': 10, 'b': 5, 'c': 3}
valid_words = {'a': 1, 'b': 1, 'd': 1}

print(subtract(book_words, valid_words))

def second_element(t):
    return t[1]
    
def print_most_common(word_counter, num=5):

    for word, freq in items[:num]:
