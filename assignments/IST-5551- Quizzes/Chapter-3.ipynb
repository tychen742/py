```python
def repeat(word, n):
    print(word * n)

spam = 'Spam, '
repeat(spam, 4)
```

```python
### first_two_lines()
def first_two_lines():
    repeat(spam, 4)
    repeat(spam, 4)
    
first_two_lines()
```

```python
### last_three_lines()
def last_three_lines():
    repeat(spam, 2)
    print('(Lovely Spam, Wonderful Spam!)')
    repeat(spam, 2)
    
last_three_lines()
```

```python
### print_verse()
def print_verse():
    first_two_lines()
    last_three_lines()
    
print_verse()
```

**What is the output of print_verse()**?
