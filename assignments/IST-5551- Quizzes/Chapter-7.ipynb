### Question 1:

What is the output of the following code? **Hint: Pay close attention to the indentation of each line in this code.**

```python
def check_letter(word, letter_to_find):
    """
    This function checks if a letter is in a word.
    """
    for letter in word:
        if letter == letter_to_find:
            return True
        else:
            return False

print(check_letter('banana', 'a'))

### Question 2:
This program tries to count all the words in a list. What is the final output?

```python
word_list = ['aa', 'aah', 'aal', 'aalii']
total = 0

for word in word_list:
    total += 1

print(total)

### Question 3:

What is the output of this code?
```python
def has_e(word):
    # .lower() converts the string to lowercase
    return 'e' in word.lower()

# Note: The word is in all caps.
print(has_e('APPLE'))

### Question 4:
This function checks if word uses only letters from the available string. What is the output of this code?

```python
def uses_only(word, available):
    """Checks whether a word uses only the available letters."""
    for letter in word:
        if letter not in available:
            return False
    return True

print(uses_only('apple', 'apl'))

### Question 5:

What is the final value of counter after this code runs?
```python
counter = 0
for i in range(4):
    counter += 1
    
counter *= 3
counter -= 2
print(counter)
