## Chapter 10 — Algorithms and Searching

### Question 1

```python
### Exercise: Binary Search Implementation
#   1. What is the Big O complexity of the function below?
#      def count_evens(nums):
#          return [n for n in nums if n % 2 == 0]
#   2. Write a function `contains(arr, target)` that uses binary search
#      (O(log n)) to check whether target is in a SORTED array.
#   3. Test with: contains([1, 3, 5, 7, 9], 5)  -> True
#                 contains([1, 3, 5, 7, 9], 4)  -> False
### Your code starts here.


### Your code ends here.
```

```python
### Solution
print('O(n)')

def contains(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return True
        if arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return False

print(contains([1, 3, 5, 7, 9], 5))
print(contains([1, 3, 5, 7, 9], 4))
```

## Chapter 11 — Exceptions and Testing

### Question 4

```python
### Exercise: Catching Exceptions
#   Write `safe_int(s)` that:
#   1. Tries to convert the string `s` to an integer with `int(s)`.
#   2. Returns the integer if conversion succeeds.
#   3. If `s` cannot be converted, catches the exception, prints a descriptive
#      message, and returns None.
#   Test:
#     safe_int('42')   -> 42
#     safe_int('3.14') -> None  (prints a message)
#     safe_int('abc')  -> None  (prints a message)
### Your code starts here.


### Your code ends here.
```

```python
### Solution
def safe_int(s):
    try:
        return int(s)
    except ValueError:
        print(f"Could not convert {s!r} to an integer.")
        return None

print(safe_int('42'))
print(safe_int('3.14'))
print(safe_int('abc'))
```

### Question 5

```python
### Exercise: Multiple Exception Types
#   Write `safe_index(lst, i)` that:
#   1. Returns lst[i] if the access succeeds.
#   2. Catches IndexError (index out of range), prints a message, and returns None.
#   3. Catches TypeError (non-integer index), prints a message, and returns None.
#   Test:
#     safe_index([10, 20, 30], 1)   -> 20
#     safe_index([10, 20, 30], 9)   -> None  (IndexError)
#     safe_index([10, 20, 30], 'a') -> None  (TypeError)
### Your code starts here.


### Your code ends here.
```

```python
### Solution
def safe_index(lst, i):
    try:
        return lst[i]
    except IndexError:
        print('Index out of range.')
        return None
    except TypeError:
        print('Index must be an integer.')
        return None

print(safe_index([10, 20, 30], 1))
print(safe_index([10, 20, 30], 9))
print(safe_index([10, 20, 30], 'a'))
```

### Question 6

```python
### Exercise: Raising Exceptions
#   Write `square_root(x)` that:
#   1. Raises a `TypeError` with a descriptive message if x is not a number (int or float).
#   2. Raises a `ValueError` with a descriptive message if x < 0.
#   3. Otherwise returns x ** 0.5.
#   Test:
#     square_root(9)       -> 3.0
#     square_root(-4)      -> raises ValueError
#     square_root('nine')  -> raises TypeError
### Your code starts here.


### Your code ends here.
```

```python
### Solution
def square_root(x):
    if not isinstance(x, (int, float)):
        raise TypeError('x must be a number')
    if x < 0:
        raise ValueError('x must be nonnegative')
    return x ** 0.5

print(square_root(9))

for value in [-4, 'nine']:
    try:
        print(square_root(value))
    except (TypeError, ValueError) as error:
        print(f'{type(error).__name__}: {error}')
```

## Chapter 12 — Object-Oriented Programming

### Question 7

```python
### Exercise: Define a Class with Methods
#   Define a class `Student` with:
#   1. `__init__(self, name, grade)` — store both as instance attributes.
#   2. `__str__` — returns a string like 'Student(Alice, 88)'.
#   3. `is_passing(self)` — returns True if grade >= 60, False otherwise.
#   Test:
#     s = Student('Alice', 88)
#     print(s)              -> Student(Alice, 88)
#     print(s.is_passing()) -> True
#     print(Student('Bob', 55).is_passing()) -> False
### Your code starts here.


### Your code ends here.
```

```python
### Solution
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __str__(self):
        return f'Student({self.name}, {self.grade})'

    def is_passing(self):
        return self.grade >= 60

s = Student('Alice', 88)
print(s)
print(s.is_passing())
print(Student('Bob', 55).is_passing())
```

### Question 8

```python
### Exercise: Inheritance and Method Overriding
#   1. Define a base class `Animal` with:
#        - `__init__(self, name)` — store name.
#        - `speak(self)` — returns the string '{name} makes a sound.'.
#   2. Define `Dog(Animal)` that overrides `speak` to return '{name} barks.'.
#   3. Define `Cat(Animal)` that overrides `speak` to return '{name} meows.'.
#   4. Write a function `make_noise(animals)` that iterates over a list of
#      Animal objects and prints the result of speak() for each one.
#   Test:
#     make_noise([Dog('Rex'), Cat('Whiskers'), Animal('Parrot')])
#     -> Rex barks.
#        Whiskers meows.
#        Parrot makes a sound.
### Your code starts here.


### Your code ends here.
```

```python
### Solution
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f'{self.name} makes a sound.'

class Dog(Animal):
    def speak(self):
        return f'{self.name} barks.'

class Cat(Animal):
    def speak(self):
        return f'{self.name} meows.'

def make_noise(animals):
    for animal in animals:
        print(animal.speak())

make_noise([Dog('Rex'), Cat('Whiskers'), Animal('Parrot')])
```

```python
### Exercise: Higher-Order Function
#   Write a function `apply_twice(f, x)` that returns the result of applying
#   function `f` to `x` two times.
#   Then call it with a lambda that adds 3.
#   Test:
#     apply_twice(lambda n: n + 3, 10) -> 16
### Your code starts here.


### Your code ends here.
```

```python
### Solution
def apply_twice(f, x):
    return f(f(x))

print(apply_twice(lambda n: n + 3, 10))
```
